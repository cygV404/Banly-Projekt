# Download Looker Dataset:


````python
import os
import pandas as pd
import kagglehub

print("1. Downloade TheLook eCommerce Datensatz...")
# Lädt den kompletten Datensatz herunter und gibt den Pfad zurück
path = kagglehub.dataset_download("mustafakeser4/looker-ecommerce-bigquery-dataset")
print("Download abgeschlossen! Die Dateien liegen unter:", path)

print("\n2. Lade Tabellen in den RAM...")


distribution_centers = pd.read_csv(os.path.join(path, "distribution_centers.csv"))
events               = pd.read_csv(os.path.join(path, "events.csv"))
inventory_items      = pd.read_csv(os.path.join(path, "inventory_items.csv"))
order_items          = pd.read_csv(os.path.join(path, "order_items.csv"))
orders               = pd.read_csv(os.path.join(path, "orders.csv"))
products             = pd.read_csv(os.path.join(path, "products.csv"))
users                = pd.read_csv(os.path.join(path, "users.csv"))

print("Datensätze wurden erfolgreich geladen!")

print("\n3. Zeilenübersicht (Kontroll-Check):")
print(f"Verteilzentren (distribution_centers): {len(distribution_centers)} Zeilen")
print(f"Events (events):                      {len(events)} Zeilen")
print(f"Inventar (inventory_items):           {len(inventory_items)} Zeilen")
print(f"Bestellpositionen (order_items):      {len(order_items)} Zeilen (Faktentabelle!)")
print(f"Bestellungen (orders):                {len(orders)} Zeilen")
print(f"Produkte (products):                  {len(products)} Zeilen")
print(f"Kunden (users):                       {len(users)} Zeilen")
````

````python


# 1. Spalte in ein echtes Datumsformat (Datetime) umwandeln
# (errors='coerce' fängt eventuelle Formatierungsfehler ab und wandelt sie in NaN um)
orders['created_at'] = pd.to_datetime(orders['created_at'], errors='coerce')

# 2. Min und Max Datum ermitteln
min_datum = orders['created_at'].min()
max_datum = orders['created_at'].max()

# 3. Schön formatiert ausgeben
print(f"Ältestes Bestelldatum (Min): {min_datum}")
print(f"Neuestes Bestelldatum (Max): {max_datum}")
print(f"Zeitspanne des Datensatzes:  {max_datum - min_datum}")
````
