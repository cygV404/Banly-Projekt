# Download Olist Dataset:


````python
import kagglehub

# Download des kompletten Olist-Datensatz
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")
print("Download abgeschlossen! Die Dateien liegen unter:", path)
````

# Alle 9 Datensätze in den RAM laden:

````python


import os
import pandas as pd


print("Lade Olist-Datensatz...")


customers         = pd.read_csv(os.path.join(path, "olist_customers_dataset.csv"))
geolocation       = pd.read_csv(os.path.join(path, "olist_geolocation_dataset.csv"))
order_items       = pd.read_csv(os.path.join(path, "olist_order_items_dataset.csv"))
order_payments    = pd.read_csv(os.path.join(path, "olist_order_payments_dataset.csv"))
order_reviews     = pd.read_csv(os.path.join(path, "olist_order_reviews_dataset.csv"))
orders            = pd.read_csv(os.path.join(path, "olist_orders_dataset.csv"))
products          = pd.read_csv(os.path.join(path, "olist_products_dataset.csv"))
sellers           = pd.read_csv(os.path.join(path, "olist_sellers_dataset.csv"))
translation       = pd.read_csv(os.path.join(path, "product_category_name_translation.csv"))

print("Alle 9 Datensätze wurden erfolgreich geladen!")


print(f"Kunden: {len(customers)} Zeilen")
print(f"Geolokation: {len(geolocation)} Zeilen")
print(f"Bestellpositionen: {len(order_items)} Zeilen")
print(f"Zahlungen: {len(order_payments)} Zeilen")
print(f"Bewertungen: {len(order_reviews)} Zeilen")
print(f"Bestellungen: {len(orders)} Zeilen")
print(f"Produkte: {len(products)} Zeilen")
print(f"Verkäufer: {len(sellers)} Zeilen")
print(f"Übersetzung: {len(translation)} Zeilen")
````
